```python
def abs_sobel_thresh(img, orient='x', sobel_kernel=3, thresh=(0, 255), channel='gray'):
    # Calculate directional gradient
    # 1) Convert to grayscale
    if channel == 'gray':
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    if channel == 'h':
        hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        gray = hls[:,:,0]
    if channel == 'l':
        hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        gray = hls[:,:,1]
    if channel == 's':
        hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        gray = hls[:,:,2]
    # 2) Take the derivative in x or y given orient = 'x' or 'y'
    if orient == 'x':
        grad = cv2.Sobel(gray, cv2.CV_64F, 1 , 0, ksize = sobel_kernel)
    if orient =='y':
        grad = cv2.Sobel(gray, cv2.CV_64F, 0 , 1, ksize = sobel_kernel)
    # 3) Take the absolute value of the derivative or gradient
    abs_grad = np.absolute(grad)
    # 4) Scale to 8-bit (0 - 255) then convert to type = np.uint8
    scale_abs_grad = np.uint8(255 * abs_grad / np.max(abs_grad))
    # 5) Create a mask of 1's where the scaled gradient magnitude 
            # is > thresh_min and < thresh_max
    # Apply threshold
    grad_binary = np.zeros_like(scale_abs_grad)
    grad_binary[(scale_abs_grad > thresh[0]) & (scale_abs_grad < thresh[1])] = 1
    return grad_binary
```

```python
def color_thresh(img, thresh=(0, 255), channel='gray'):
    # Calculate gradient direction
    # 1) Convert to grayscale
    #gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    if channel == 'gray':
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    if channel == 'h':
        hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        gray = hls[:,:,0]
    if channel == 'l':
        hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        gray = hls[:,:,1]
    if channel == 's':
        hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        gray = hls[:,:,2]
    if channel == 'v':
        luv = cv2.cvtColor(img, cv2.COLOR_RGB2LUV)
        gray = luv[:,:,2]
    # Apply threshold
    # 5) Create a binary mask where direction thresholds are met
    dir_binary = np.zeros_like(gray)
    # 6) Return this mask as your binary_output image
    dir_binary[(gray>thresh[0]) & (gray<thresh[1])] = 1
    
    return dir_binary
```

```python
def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
```

Try different combinations and see what you get. For example, here is a selection for pixels where both the x and y gradients meet the threshold criteria, or the gradient magnitude and direction are both within their threshold values.

```python
%matplotlib inline

# Read an image
f_name = 'test5.jpg'
image = mpimg.imread('results/undst/' + f_name)

# Apply each of the thresholding functions
yellow_line = color_thresh(image, thresh=(160, 255), channel='v')
#yellow_line = color_thresh(image, thresh=(18, 25), channel='h')
#white_line = color_thresh(image, thresh=(212, 255), channel='gray')
white_line = color_thresh(image, thresh=(205, 255), channel='l')
#gradx = abs_sobel_thresh(image, orient='x', sobel_kernel=3, thresh=(30, 100), channel = 'l')
gradx = abs_sobel_thresh(image, orient='x', sobel_kernel=3, thresh=(30, 80), channel = 'gray')

# Combine all in one 
binary = np.zeros_like(gradx)
binary[(yellow_line == 1) | (white_line == 1) | (gradx == 1)] = 1

# Stack in RGB format to visualization
color_binary = np.dstack(( gradx*255, yellow_line*255, white_line*255))

# Plot the result
f, ((ax1), (ax2), (ax3)) = plt.subplots(3, 1, figsize=(20, 8))
f.tight_layout()
ax1.set_title('Input Image', fontsize=20)
ax1.imshow(image)
ax2.set_title('Color binary Image', fontsize=20)
ax2.imshow(color_binary)
ax3.set_title('Binary Image', fontsize=20)
ax3.imshow(binary, cmap='gray')
plt.subplots_adjust(left=0., right=1, top=1.2, bottom=0.)

# Save the binary image
cv2.imwrite('results/binary_image/' + f_name, binary)
```

```python
dist_limit = 410
imshape = image.shape
vertices = np.array([[(80,imshape[0]),(600, dist_limit), 
                      (680, dist_limit), (1200,imshape[0])]], dtype=np.int32)

masked_binary = region_of_interest(binary,vertices)

f = plt.figure()
#plt.plot(data)
f.suptitle('Masked binary', fontsize=20)
#f.title('Input Image', fontsize=20)
plt.imshow(masked_binary, cmap='Greys_r')
```
